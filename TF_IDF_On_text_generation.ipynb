```python
# prompt: given phrase generate paragraph using any model

from transformers import pipeline

generator = pipeline("text-generation")
```

```python

phrase = "Machine learning algorithms are used for data analysis and pattern recognition."

# Generate multiple documents with different fields
fields = ["Introduction", "Applications", "Challenges", "Conclusion"]
documents = {}

for field in fields:
    result = generator(phrase + f" {field}.", max_length=200, num_return_sequences=1)
    documents[field] = result[0]["generated_text"]

# Print the generated documents
for field, text in documents.items():
    print(f"{field}:")
    print(text)
    print("-" * 50)
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
from nltk.stem import PorterStemmer

def normalize_text(text):
    # Convert text to lowercase
    return text.lower()

def tokenize_text(text):
    # Tokenize text into words
    return word_tokenize(text)

def lemmatizing(word):
    lemmatizer = WordNetLemmatizer()
    return lemmatizer.lemmatize(word)

# Process documents
def preprocessing(text):
    # Step 1: Remove Charchters Excepts Letters
    preprocessed_text = re.sub(r'\s+', ' ', text)
    preprocessed_text = re.sub(r'[^a-zA-Z\s]', '',preprocessed_text)
    # Step 2: Normalization
    normalized_doc = normalize_text(preprocessed_text)
    # Step 3: Tokenization
    tokens = tokenize_text(normalized_doc)

    # Step 4: Remove stop words
    stop_words = stopwords.words('english')

    lemmatize_words = [lemmatizing(word) for word in tokens if word not in stop_words] # Lemmatizing

    return lemmatize_words
```

```python
docs = []
for field, text in documents.items():
    docs.append(''.join(text))
docs
```

```python
# Calculate TF-IDF
vectorizer = TfidfVectorizer(tokenizer=preprocessing)
tfidf_matrix = vectorizer.fit_transform(docs)
feature_names = vectorizer.get_feature_names_out()
c=0
# Print TF-IDF values for each document
for i, field in enumerate(fields):
    print(f"TF-IDF values for {field}:")
    for j, feature in enumerate(feature_names):
        if tfidf_matrix[i, j] > 0.0:
          c+=1
          print(f"{feature}: {tfidf_matrix[i, j]}")
    print(c)
    c=0
    print("-" * 50)
```

```python
#Importing required module
import numpy as np
from nltk.tokenize import  word_tokenize

# Process documents
processed_documents = []
for doc in docs:
  pro = preprocessing(doc)
  processed_documents.append(pro)
#Preprocessing the text data
sentences = []
word_set = []

for sent in processed_documents:
    sentences.append(sent)
    for word in sent:
        word_set.append(word)

#Set of vocab
word_set = set(word_set)
#Total documents in our corpus
total_documents = len(sentences)

#Creating an index for each word in our vocab.
index_dict = {} #Dictionary to store index for each word
i = 0
for word in word_set:
    index_dict[word] = i
    i += 1
```

```python
#Create a count dictionary

def count_dict(sentences):
    word_count = {}
    for word in word_set:
        word_count[word] = 0
        for sent in sentences:
            if word in sent:
                word_count[word] += 1
    return word_count

word_count = count_dict(sentences)
```

```python
#Term Frequency
def termfreq(document, word):
    N = len(document)
    occurance = len([token for token in document if token == word])
    return occurance/N
```

```python
#Inverse Document Frequency

def inverse_doc_freq(word):

    word_occurance = word_count[word]

    return np.log((total_documents + 1)/(word_occurance + 1)) + 1
```

```python
def tf_idf(sentence):
    tf_idf_vec = np.zeros((len(word_set),))
    for word in sentence:
        tf = termfreq(sentence,word)
        idf = inverse_doc_freq(word)

        value = tf*idf
        tf_idf_vec[index_dict[word]] = value

    return tf_idf_vec
```

```python
#TF-IDF Encoded text corpus
vectors = []
for sent in sentences:
    vec = tf_idf(sent)
    vectors.append(vec)

# Normalize TF-IDF vector
tf_idf_vec_norm = np.array(vectors) / np.linalg.norm(vectors, axis=1, keepdims=True)


c=0
arr = []
# Print TF-IDF values for each document
for i, field in enumerate(fields):
    print(f"TF-IDF values for {field}:")
    for j, feature in enumerate(word_set):
        if vectors[i][j] > 0.0:
          c+=1
          arr.append([feature,tf_idf_vec_norm[i][j]])
    # Sort arr based on feature names
    arr = sorted(arr, key=lambda x: x[0])
    for item in arr:
        print(f"{item[0]}: {item[1]}")
    print(c)
    arr = []
    c=0
    print("-" * 50)


```
